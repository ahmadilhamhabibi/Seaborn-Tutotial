````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
sns.set_style('darkgrid')

plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('axes', labelsize=14)
````

````python
sns.__version__
````

````python
penguins = sns.load_dataset("penguins")

penguins.head()
````

````python
penguins.shape
````

````python
penguins.dropna(inplace=True)
````

````python
penguins.shape
````

## Basic Histplot

````python
sns.histplot(penguins.bill_length_mm)
````

````python
sns.histplot(data=penguins, y="bill_length_mm")
````

### ```kde```

````python
sns.histplot(penguins.bill_length_mm, kde=True)
````

### ```bins```

````python
sns.histplot(penguins.bill_length_mm, bins=20)
````

````python
sns.histplot(penguins.bill_length_mm, bins=[30, 35, 40, 45, 50, 55, 60])
````

### ```binwidth``` and ```binrange```

````python
sns.histplot(penguins.bill_length_mm, binwidth=5)
````

````python
sns.histplot(penguins.bill_length_mm, binrange=(30, 60))
````

````python
sns.histplot(penguins.bill_length_mm, binwidth=5, binrange=(30, 60))
````

### ``cumulative``

````python
sns.histplot(penguins.bill_length_mm, cumulative=True,
             fill=False, element='step')
````

## Categorical Variables

### ``hue``

````python
sns.histplot(data=penguins, x="bill_length_mm", hue="species")
````

### ``element``

````python
sns.histplot(data=penguins, x="bill_length_mm", 
             hue="species", element="step")
````

````python
sns.histplot(data=penguins, x="bill_length_mm", 
             hue="species", element="poly")
````

### ``multiple``

````python
sns.histplot(data=penguins, x="bill_length_mm",
             hue="species",
             multiple="stack")
````

````python
sns.histplot(data=penguins, x="bill_length_mm",
             hue="species",
             multiple="fill")
````

## Bivariate Histplot

````python
sns.set_style("white")
````

````python
sns.histplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
````

### ``cbar``

````python
sns.histplot(data=penguins, x="bill_length_mm", y="bill_depth_mm",
             cbar=True)
````

### ``hue``

````python
sns.histplot(data=penguins, x="bill_length_mm", y="bill_depth_mm",
             hue="species")
````

## Styling

## ``shrink``

````python
sns.histplot(data= penguins, x="species", hue="sex",
             multiple="dodge", shrink=0.8)
````

## ``color``, ``palette``

````python
sns.histplot(penguins.bill_length_mm, color="tomato")
````

````python
sns.histplot(data= penguins, x="species", hue="sex",
             multiple="dodge", shrink=0.8,
             palette="bone")
````

## ``fill``

````python
sns.histplot(penguins.bill_length_mm, fill=False)
````
