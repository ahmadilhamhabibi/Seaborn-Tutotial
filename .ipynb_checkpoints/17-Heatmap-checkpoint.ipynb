# Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cars = sns.load_dataset("mpg")
```

```python
cars.head()
```

## Basic

```python
cars.groupby("origin").cylinders.value_counts()
```

```python
cars.groupby("origin").cylinders.value_counts().unstack().fillna(0)
```

```python
origin_cyl = cars.groupby("origin").cylinders.value_counts().unstack().fillna(0)
```

```python
sns.heatmap(origin_cyl)
```

### Arrays

```python
test_array = np.array([[1, 2], [3, 4]])

test_array
```

```python
sns.heatmap(test_array)
```

### Transpose

```python
origin_cyl.T
```

```python
sns.heatmap(origin_cyl.T)
```

### Correlation

```python
cars.corr()
```

```python
sns.heatmap(cars.corr())
```

### Color Palette

Seaborn offers 170 different color palette options.

We'll start with a SEQUENTIAL palette.

```python
sns.heatmap(origin_cyl, cmap="Blues");
```

Now let's try a DIVERGING palette.

```python
sns.heatmap(cars.corr(), cmap='RdBu')
```

### Controling the Colorbar Range: center, vmax, vmin

```python
sns.heatmap(cars.corr(), cmap="RdBu", center=0)
```

```python
sns.heatmap(cars.corr(), cmap="RdBu", vmin=-1, vmax=1)
```

### Annotations

```python
sns.heatmap(origin_cyl, cmap="Blues", annot=True)
```

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True, fmt=".0f");  #fmt=".1f" for one decimal place
```

```python
sns.heatmap(origin_cyl, 
            cmap='Blues', 
            annot=True, 
            fmt=".0f",
            annot_kws={
                'fontsize': 12,
                'fontweight': 'bold',
                'fontfamily': 'serif',
               # 'color': 'black'
            }
           )
```

## Styling

### linewidth

```python
sns.heatmap(cars.corr(), cmap="RdBu", linewidths=1)
```

### linecolor

```python
sns.heatmap(cars.corr(), cmap="RdBu", linewidths=1, linecolor="black")
```

### Tick Labels

```python
sns.heatmap(cars.corr(), cmap="RdBu")
```

```python
labels = ['MPG', 'Cylinders', 'Displacement', 'Horsepower',
          'Weight', 'Acceleration', 'Model Year'
         ]
```

```python
sns.heatmap(cars.corr(), 
            cmap="RdBu",
            xticklabels=labels,
            yticklabels=labels
           )
```

```python
sns.heatmap(cars.corr(), 
            cmap="RdBu",
            xticklabels=labels,
            yticklabels=False
           )
```

### square

```python
sns.heatmap(cars.corr(), cmap="RdBu", square=True)
```
