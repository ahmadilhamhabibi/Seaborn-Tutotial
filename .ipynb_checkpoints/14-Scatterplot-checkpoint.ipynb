# Scatterplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg").dropna()

cars.shape
```

```python
cars.head()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])].sample(n=100, random_state=24)

cars.shape
```

## Basic

```python
sns.set_style("dark")
```

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg")
```

```python
sns.scatterplot(data=cars, x="horsepower", y="acceleration")
```

## Semantic Variables

```python
sns.set_style("darkgrid")
```

### hue

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", hue="origin")
```

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", hue="origin",
                palette="mako")
```

### style

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", style="origin",
                palette="mako")
```

## size

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", size="origin")
```

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", 
                size="origin", sizes=[50, 100, 150])
```

### hue + style + size

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", 
                hue="origin",
                style="origin",
                size="origin",
                sizes=[50, 100, 150])
```

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", 
                hue="cylinders",
                style="cylinders",
                size="cylinders",
                sizes=[50, 100, 150])
```

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", 
                hue="origin",
                style="cylinders",
                size="cylinders",
                sizes=[50, 100, 150])
```

## Additional Styling

### alpha

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", hue="origin",
                s=100, alpha=0.6)
```

### marker styling

Check the matplotlib scatter and the marker style docs for more styling of the markers themselves

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", 
                s=200, marker="*")
```

```python
sns.scatterplot(data=cars, x="horsepower", y="mpg", 
                s=200, marker="*", edgecolor="black")
```
