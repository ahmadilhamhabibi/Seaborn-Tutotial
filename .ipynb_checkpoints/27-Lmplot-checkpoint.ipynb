# Lmplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins").dropna()

df.head()
```

## Basics

```python
sns.set_style("white")
```

```python
sns.lmplot(data=df, x="bill_length_mm", y="bill_depth_mm")
```

### ``hue``

```python
sns.lmplot(data=df, x="bill_length_mm", y="bill_depth_mm",
           hue="species"
          )
```

### ``markers``

```python
sns.lmplot(data=df, x="bill_length_mm", y="bill_depth_mm",
           hue="species",
           markers=["o", "x", "d"]
          )
```

## Small Multiples (FacetGrid)

### ``col``

```python
sns.lmplot(data=df, x="bill_length_mm", y="bill_depth_mm",
           hue="species",
           col="species"
          )
```

### ``row``

```python
sns.lmplot(data=df, x="bill_length_mm", y="bill_depth_mm",
           hue="species",
           col="species",
           row="sex",
           truncate=False
          )
```

## Extras

### ``x_bins``

```python
sns.lmplot(x="bill_depth_mm", y="bill_length_mm",
           data=df[df.species=="Chinstrap"],
           x_bins=5
          )
```

### Other models: Logistic Regression, ``logistic``

```python
df['sex_binary'] = [(x == 'Male') for x in df.sex]
```

```python
df.head()
```

```python
sns.lmplot(data=df, x="body_mass_g", y="sex_binary", logistic=True, n_boot=100)
```

```python
sns.lmplot(data=df, x="body_mass_g", y="sex_binary", logistic=True, n_boot=100,
           hue="species", col="species")
```

## Styling

### ``scatter_kws``

```python
sns.lmplot(data=df, x="bill_depth_mm", y="bill_length_mm", 
           hue="species",
           col="species",
           scatter_kws={"s": 150, "edgecolor": "black"}
          )
```

### ``line_kws``

```python
sns.lmplot(data=df, x="bill_depth_mm", y="bill_length_mm", 
           hue="species",
           col="species",
           line_kws={"lw": 5, "ls": "-.", "color": "black"}
          )
```

### FacetGrid Object

```python
g = sns.lmplot(data=df, x="bill_depth_mm", y="bill_length_mm", 
           hue="species",
           col="species",
          )
```

```python
type(g)
```

```python
g = sns.lmplot(data=df, x="bill_depth_mm", y="bill_length_mm", 
           hue="species",
           col="species",
          )

g.set_titles(col_template="{col_name} Penguins")
g.set_xlabels("Bill Depth (mm)")
```

## Example

```python
sns.set(style="white", font_scale=1.7)
g = sns.lmplot(data=df, x="bill_length_mm", y="bill_depth_mm",  
           hue="species",
           col="species",
           truncate=False
          )

g.set_titles(col_template="{col_name} Penguins")
g.set_xlabels("Bill Length (mm)")
g.set_ylabels("Bill Depth (mm)")
plt.tight_layout()
```
