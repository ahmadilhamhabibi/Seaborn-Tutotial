# Swarmplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cars = sns.load_dataset("mpg").dropna()

cars.shape
```

```python
cars.head()
```

## Basic

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.swarmplot(data=cars, x="horsepower")
```

```python
sns.swarmplot(data=cars, x="origin", y="horsepower")
```

```python
# sns.stripplot(data=cars, x="origin", y="horsepower")
```

```python
sns.swarmplot(data=cars, x="origin", y="horsepower", hue="cylinders")
```

```python
sns.swarmplot(data=cars, x="origin", y="horsepower", hue="cylinders",
              dodge=True)
```

## Swarmplot Overlay

```python
usa = cars[cars.origin=="usa"]
```

### With Boxplot

```python
sns.boxplot(data=usa, x="cylinders", y="horsepower")
sns.swarmplot(data=usa, x="cylinders", y="horsepower")
```

```python
sns.boxplot(data=usa, x="cylinders", y="horsepower", whis=np.inf)
sns.swarmplot(data=usa, x="cylinders", y="horsepower", 
              color="black", alpha=0.5)
```

### With Violinplot

```python
sns.violinplot(data=usa, x="cylinders", y="horsepower")
sns.swarmplot(data=usa, x="cylinders", y="horsepower")
```

```python
sns.violinplot(data=usa, x="cylinders", y="horsepower", scale="width")
sns.swarmplot(data=usa, x="cylinders", y="horsepower",
              color="white", size=4)
```

```python
sns.violinplot(data=usa, x="cylinders", y="horsepower", scale="width",
               inner=None, color='xkcd:forest green')
sns.swarmplot(data=usa, x="cylinders", y="horsepower",
              color="white", size=3)
```

## Styling

### linewidth, edgecolor

```python
sns.swarmplot(data=cars, x="horsepower", linewidth=1)
```

```python
sns.swarmplot(data=cars, x="horsepower", linewidth=1, edgecolor="black")
```

### Matplotlib Scatter Properties: marker

Check the matplotlib scatter and the marker style docs for more styling of the markers themselves

```python
plt.figure(figsize=(8,4))

sns.swarmplot(data=cars, x="horsepower", marker="*")
```
