## Basics

```python
sns.set_style("darkgrid")
```

```python
sns.jointplot(data=geyser, x="waiting", y="duration")
```

## Plot Kinds

There are currently six kinds of jointplots allowed in seaborn:

1. scatter (default)
2. kde
3. reg
4. hist
5. hex
6. resid

### scatter (Default)

```python
sns.jointplot(data=geyser, x="waiting", y="duration",
              kind="scatter")
```

### kde

```python
sns.jointplot(data=geyser, x="waiting", y="duration",
              kind="kde")
```

### reg

```python
sns.jointplot(data=geyser, x="waiting", y="duration",
              kind="reg")
```

### hist

```python
sns.jointplot(data=geyser, x="waiting", y="duration",
              kind="hist")
```

## Return JointGrid

```python
g = sns.jointplot(data=geyser, x="waiting", y="duration")
```

```python
type(g)
```

```python
g.plot_joint(sns.kdeplot, color="gray", levels=5)
```

## Styling

### color, palette

```python
sns.jointplot(data=geyser, x="waiting", y="duration", 
              color="green")
```

```python
sns.jointplot(data=geyser, x="waiting", y="duration", 
              hue="kind",
              palette="autumn")
```

### height, space, ratio

```python
sns.jointplot(data=geyser, x="waiting", y="duration", 
              height=8,   #Default: 6
              space=0.1,  #Default: 0.2
              ratio=2     #Default: 5
             );
```

### joint_kws

Joint keywords should match the figure type in the middle of the jointplot.

```python
sns.jointplot(data=geyser, x="waiting", y="duration", 
              joint_kws={"marker": 5})
```

```python
sns.jointplot(data=geyser, x="waiting", y="duration", 
              kind="reg",
              joint_kws={"ci": None, "order": 2})
```

### marginal_kws

Marginal keywords should match the type of plot for the marginal distribution plots (the outside plots).

```python
sns.jointplot(data=geyser, x="waiting", y="duration",
              marginal_kws={"color": "xkcd:golden"})
```

```python
sns.jointplot(data=geyser, x="waiting", y="duration",
              kind="kde",
              marginal_kws={"lw": 4, "shade": True})
```
