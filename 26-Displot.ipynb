# Displot

```python
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg").dropna()

cars.head()
```

## Basics

```python
sns.set_style("white")
```

```python
sns.displot(data=cars, x="weight")
```

### ``kde``

```python
sns.displot(data=cars, x="weight", kde=True)
```

### ``kind``

#### **Kernel Density Estimation Plot:** ``kind='kde'``

```python
sns.displot(data=cars, x="weight", kind="kde")
```

#### **Empirical Cumulative Distribution Function Plot:** ``kind='ecdf'``

```python
sns.displot(data=cars, x="weight", kind="ecdf")
```

### ``rug``

```python
sns.displot(data=cars, x="weight", kind="kde", rug=True)
```

## Categorical Variables: ``hue``

```python
sns.displot(data=cars, x="weight", hue="origin")
```

```python
sns.displot(data=cars, x="weight", kind="kde", hue="origin")
```

## Bivariate (2D) Options

```python
sns.displot(data=cars, x="weight", y="horsepower")
```

```python
sns.displot(data=cars, x="weight", y="horsepower", kind="kde")
```

```python
sns.displot(data=cars, x="weight", y="horsepower", kind="kde", rug=True)
```

## Small Multiples (FacetGrid)

```python
sns.displot(data=cars, x="weight", hue="origin", col="origin")
```

```python
sns.displot(data=cars, x="weight", col="origin", kind="kde", rug=True)
```

```python
sns.displot(x='weight', data=cars[cars.cylinders.isin([4, 6, 8])], 
            col='origin', 
            row='cylinders', 
            hue='cylinders'
           );
```

## Styling

### ``palette``

```python
sns.displot(data=cars, x="weight", hue="origin", col="origin", palette="crest")
```

### ``height``, ``aspect``

```python
sns.displot(data=cars, x="weight", col="origin", height=2)
```

```python
sns.displot(data=cars, x="weight", col="origin", height=2, aspect=1.5)
```

### Other Keywords: Plot Specific

In general, additional keyword arguments will be passed to the kind of plot selected.

```python
sns.displot(data=cars, x="weight", bins=20)
```

```python
sns.displot(data=cars, x="weight", kind="kde", fill=True, lw=3)
```

### ``rug_kws``

```python
sns.displot(data=cars, x="weight", kind="kde", rug=True, rug_kws={"height": 0.05})
```

### FacetGrid Styling

#### ``facet_kws``

```python
sns.displot(data=cars, x="weight", col="origin", facet_kws={"despine": False})
```

### FacetGrid Object

```python
g = sns.displot(data=cars, x="weight", col="origin")
```

```python
type(g)
```

```python
g.axes_dict.items()
```

```python
g = sns.displot(data=cars, x="weight", col="origin")

for col_val, ax in g.axes_dict.items():
    if col_val == "europe":
        ax.set_facecolor("#E2F2FF")
```
